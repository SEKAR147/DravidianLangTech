```python
from ctransformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained(
    "tensorblock/abhinand_tamil-llama-7b-instruct-v0.1-GGUF",
    model_type="llama",
    gpu_layers=0,
    max_new_tokens=64,
    temperature=0.1
)
```

```python
def classify_sentiment_tamil(text):
    prompt = f"""
You are a sentiment classification system.

Classify the sentiment of the following Tamil sentence into ONLY one:
Positive, Negative, or Neutral.

Sentence:
{text}

Answer with only one word.
"""

    output = model(prompt)
    return output.strip()
```

```python
text = "இந்த படம் மிகவும் அருமையாக உள்ளது"
print(classify_sentiment_tamil(text))
```
